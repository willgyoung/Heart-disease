```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
hd = pd.read_csv('heart.csv')
```

```python
hd.head()
```

```python
hd = hd.rename(columns = {'cp':'chest_pain', 'trestbps':'resting_bp', 'chol':'cholesterol', 'fbs':'fasting_bs', 
                         'thalach':'max_hr', 'exang':'exercise_angina', 'oldpeak':'ST_depression', 'thal':'defect', 
                         'target':'CHD_present', 'ca':'vessels_colored'})
hd.head()
```

```python
hd = hd.rename(columns = {'slope':'ST_slope'})
```

```python
hd.info()
```

```python
hd['sex'] = np.where(hd['sex'] == 1, 'M', 'F')
hd.sex.head()
```

```python
# The person who originally made this dataset flipped those that did and didn't have heart disease. 
# This was something I only realized after working with the data for a bit. This is the code
# to flip those that do and don't have CHD. 
hd['CHD_present'] = np.where(hd['CHD_present'] == 1, 0, 1)
hd.CHD_present.head()
```

```python
hd['chest_pain'] = pd.Categorical(hd['chest_pain'])
hd['fasting_bs'] = hd['fasting_bs'].astype(bool)
hd['restecg'] = pd.Categorical(hd['restecg'])
hd['exercise_angina'] = hd['exercise_angina'].astype(bool)
hd['ST_slope'] = pd.Categorical(hd['ST_slope'])
hd['defect'] = pd.Categorical(hd['defect'])
hd['CHD_present'] = hd['CHD_present'].astype(bool)
```

```python
hd.info()
```

```python
hd.describe()
```

#### Since the description for resting ECG is unclear, I decided to drop it 

```python
hd = hd.drop(columns = ['restecg'])
```

```python
hd.head()
```

```python
chest_pain_types = []
for row in hd['chest_pain']:
    if row == 1:
        chest_pain_types.append('typical_angina')
    elif row == 2:
        chest_pain_types.append('atypical_angina')
    elif row == 3:
        chest_pain_types.append('non-aginal_pain')
    else:
        chest_pain_types.append('asymptomatic')
        
len(chest_pain_types)
```

```python
hd['chest_pain'] = chest_pain_types
```

```python
hd.head()
```

```python
slope_type = []
for row in hd['ST_slope']:
    if row == 0:
        slope_type.append('upsloping')
    elif row == 1:
        slope_type.append('flat')
    else:
        slope_type.append('downsloping')

        
print(len(slope_type))

hd['ST_slope'] = slope_type
hd.head()
```

```python
defect_type = []
for row in hd['defect']:
    if row == 1:
        defect_type.append('normal')
    elif row == 2:
        defect_type.append('fixed_defect')
    elif row == 3:
        defect_type.append('reversable_defect')
    else:
        defect_type.append('NaN')
        
hd['defect'] = defect_type
hd.head()
```

```python
hd.isnull().sum()
```

### First thing is to inspect and see if there are any strong correlations between heart disease and the other factors

```python
plt.figure(figsize = (20, 10))
sns.heatmap(hd.corr(), annot = True)
plt.show()
```

### From just looking at the correlations the four greatest indicators of heart disease are the maximum heart rate achieved (inverted relationship), whether or not the patient experienced exercise angina, the level of ST wave depression, and the number of vessels colored in the fluoroscopy. I'm also going to look and some of the other data points and see what kind of covariance there is between them.
